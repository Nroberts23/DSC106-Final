```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def conv_to_levels(l):
    if (l <= 35):
        return 0
    elif (l <= 104):
        return 1
    return 2

def level_mean(l):
    if (l == 0):
        return 'safe'
    if (l == 1):
        return 'warning'
    return 'dangerous'
```

```python
df = pd.read_csv('San_Diego.csv')
#df['danger_levels'] = df['enterobacteria'].apply(level_mean)
df.head()
```

```python
df['date'] = pd.to_datetime(df.date)
```

```python
grouped = df.groupby('siteID')
```

```python
d_levs = pd.DataFrame()
location = []
no_rain = []
rain = []

for g in grouped:
    location.append(g[0])
    frame = g[1]
    g_grouped = frame.groupby('raininlast3days')
    for roup in g_grouped:
        roup_frame = roup[1]
        if (roup[0] == 'no'):
            no_rain.append(np.mean(roup[1].enterobacteria))
        else:
            rain.append(np.mean(roup[1].enterobacteria))
            
d_levs['location'] = location
d_levs['no_rain'] = no_rain
d_levs['rain'] = rain

d_levs = d_levs.set_index('location')
```

```python
d_levs.to_csv('average_ent_levels.csv')
```
